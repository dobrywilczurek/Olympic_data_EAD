```python
sns.countplot(data=data,x="Year",hue="Medal")
plt.title("Medal Distribution over Years")
plt.xticks(rotation=45)
plt.show()
```

```python
year_avg_age = data.groupby("Year")["Age"].mean()
year_avg_age
```

```python
sport_median_height = data.groupby("Sport")["Height"].median()
sport_median_height
```

```python
country_gender_count = data.groupby(["NOC", "Sex"])["ID"].count()
country_gender_count
```

```python
country_gold_medals = data[data["Medal"] == "Gold"].groupby("NOC")["Medal"].count()
country_gold_medals
```

```python
sport_gender_avg_weight = data.groupby(["Sport","Sex"])["Weight"].mean()
sport_gender_avg_weight
```

```python
sport_event_count = data.groupby("Sport")["Event"].nunique().sort_values(ascending=False)
sport_event_count
```

```python
sport_event_count.plot(kind="bar",figsize=(20,10))
plt.title("Numer of unique events per sport")
plt.xticks(rotation=45)
plt.show()
```

```python
tallest_male_atlh = data[data["Sex"]=="M"].groupby('Name')['Height'].mean().sort_values(ascending=False)
tallest_male_atlh
```

```python
tallest_female_atlh = data.loc[data["Sex"]=="F"].groupby('Name')['Height'].mean().sort_values(ascending=False)
tallest_female_atlh
```

```python
athletes_per_year = data.groupby("Year")["ID"].nunique().reset_index()
sns.lineplot(data=athletes_per_year, x="Year", y="ID")
plt.title("Unique athletes per years")
plt.xlabel("Year")
plt.ylabel("Number of athletes")
plt.show()
```

```python
country_avg_age = data.groupby("NOC")["Age"].mean().sort_values(ascending=False)
country_avg_age
```

```python
country_avg_age.head(10).plot(kind='bar', figsize=(20,10))
plt.xticks(rotation=45)
plt.title("Top 10 countries with the highest average age")
plt.xlabel("Coutry")
plt.ylabel("Avegare age")
plt.show()
```

```python
sns.boxplot(data=data, x="Season", y="Age")
plt.title("Age Distribution by Season")
plt.xlabel("Season")
plt.ylabel("Age")
plt.show()
```

```python
sns.violinplot(data=data, x="Medal", y="Height", palette="Set3")
plt.title("Distribution of Height by Medals")
plt.xlabel("Medal")
plt.ylabel("Height")
plt.show()
```

```python
df1=data.drop_duplicates(subset=["Sport", "Name"])
plt.figure(figsize=(6,12))
sns.countplot(data=df1 ,y="Sport")
plt.title("Count of Sports participants")
plt.ylabel("Sport")
plt.xlabel("Number of participants")
plt.show()
```

```python
most_medals_by_country = data.groupby("NOC")["Medal"].count()
most_medals_by_country[most_medals_by_country>200].sort_values(ascending=False)
```

```python
most_medals_by_country = most_medals_by_country[most_medals_by_country>10]
most_medals_by_country = most_medals_by_country.sort_values(ascending=False)
top_10_countries_medal = most_medals_by_country.head(10)
```

```python
top_10_countries_medal.plot(kind='bar', figsize=(20,6))
plt.title("Top 10 countries with the most Medals")
plt.xticks(rotation=45)
plt.xlabel("Country")
plt.ylabel("Total of Medals")
plt.show()
```
